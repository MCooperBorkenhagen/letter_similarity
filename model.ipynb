```python
import numpy as np
import json
from keras.layers import Input, Dense
from keras.models import Model
```

## Data
Read in the data from `letters.json`

```python
with open("letters.json") as f:
    letter_dict = json.load(f)
```

```python
flattened_data = np.array([np.array(bitmap).flatten() for bitmap in letter_dict.values()])
```

```python
letters = [letter for letter in letter_dict.keys()]
```

```python
letters
```

## Write model architecture

```python

input = Input(shape=(80,))

encoder = Dense(64, activation='relu')(input)
encoder = Dense(32, activation='relu')(encoder)
encoder = Dense(16, activation='relu')(encoder)
encoder = Dense(8, activation='relu')(encoder)
#
decoder = Dense(16, activation='relu')(encoder)
decoder = Dense(32, activation='relu')(decoder)
decoder = Dense(64, activation='relu')(decoder)
decoder = Dense(80, activation='sigmoid')(decoder)

# the model
model = Model(input, decoder)
model.compile(optimizer='adam', loss='binary_crossentropy')

# print summary
model.summary()

```

## Fit the model

```python
model.fit(flattened_data, flattened_data, epochs=300, shuffle=True)
```

## Latent representations

```python
narrowest_layer = 'dense_3'

encoder = Model(inputs=model.input, outputs=model.get_layer(narrowest_layer).output)

# Get the latent representations at the most narrow layer
latent_representations = encoder.predict(flattened_data)
```
